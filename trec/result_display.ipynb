```python
import numpy as np
import pandas as pd
import glob
import dataframe_image as dfi
```

```python
def get_data(filename):
  data = {}
  with open(filename) as f:
      for line in f:
        (k,v) = line.split()
        data[k] = v
  return data
```

```python
files=glob.glob('*.txt')
df=pd.DataFrame.from_dict(get_data(files[0]), orient='index',columns=[files[0].split('.')[0]])
print(files[0].split('.')[0])
```

```python
for i in range(1, len(files)):
  val=get_data(files[i])
  df[files[i].split('.')[0]]=val.values()

df
```
